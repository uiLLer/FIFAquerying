##### Criando uma função pra diminuir a quantidade de linhas ao abrir uma conexão, executar e visualizar uma query

```python
def executeQuery(query):
    conn = cx_Oracle.connect('system/oracle@localhost:1521/xe')
    cursor = conn.cursor()
    cursor.execute(query)
    rows = cursor.fetchall()
    columns = [col[0] for col in cursor.description]
    df = pd.DataFrame(rows, columns=columns)
    cursor.close()
    conn.close()
    return df
```

### Questões:

Dentre todos os fifas, quais são os melhores jogadores (que possuem maior RTG)?

```python
queryMaioresRTG = """
WITH 
    tabelinhas_18 AS (
        SELECT name, rtg, card, nation
        FROM (
            SELECT name, rtg, card, nation
            FROM FIFA.PLAYERS_18
            ORDER BY rtg DESC
        ) 
        WHERE rownum <= 3
    ),
    tabelinhas_19 AS (
        SELECT name, rtg, card, nation
        FROM (
            SELECT name, rtg, card, nation
            FROM FIFA.PLAYERS_19
            ORDER BY rtg DESC
        ) 
        WHERE rownum <= 3
    ),
    tabelinhas_20 AS (
        SELECT name, rtg, card, nation
        FROM (
            SELECT name, rtg, card, nation
            FROM FIFA.PLAYERS_20
            ORDER BY rtg DESC
        ) 
        WHERE rownum <= 3
    ),
    tabelinhas_21 AS (
        SELECT name, rtg, card, nation
        FROM (
            SELECT name, rtg, card, nation
            FROM FIFA.PLAYERS_21
            ORDER BY rtg DESC
        ) 
        WHERE rownum <= 3
    ),
    tabelinhas_22 AS (
        SELECT name, rtg, card, nation
        FROM (
            SELECT name, rtg, card, nation
            FROM FIFA.PLAYERS_22
            ORDER BY rtg DESC
        ) 
        WHERE rownum <= 3
    )
    SELECT DISTINCT(NAME), rtg, nation 
    FROM (
        SELECT * FROM tabelinhas_18
        UNION ALL
        SELECT * FROM tabelinhas_19
        UNION ALL
        SELECT * FROM tabelinhas_20
        UNION ALL
        SELECT * FROM tabelinhas_21
        UNION ALL
        SELECT * FROM tabelinhas_22
    )
"""
executeQuery(queryMaioresRTG)
```

Qual a Nação com maior média de SHO, DRI e DEF

```python
templateQueryMedia = """
SELECT
    NATIONALITY,
    ROUND((AVERAGE_SHOOTING + AVERAGE_DRIBBLE + AVERAGE_DEFENSE) / 3, 2) AS TOTAL_AVERAGE
FROM (
    SELECT
        NATION AS NATIONALITY,
        AVG(SHO) AS AVERAGE_SHOOTING,
        AVG(DRI) AS AVERAGE_DRIBBLE,
        AVG(DEF) AS AVERAGE_DEFENSE
    FROM FIFA.PLAYERS_18
    GROUP BY NATION
) AVG_STATS
order by TOTAL_AVERAGE desc
"""

executeQuery(templateQueryMedia)
```
